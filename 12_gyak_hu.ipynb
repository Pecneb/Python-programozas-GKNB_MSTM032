## Gyakorlás: FIFA'19 játékosok

A [fifa19.csv.gz](fifa19.csv.gz) fájl a FIFA'19 játékban szereplő focistákról tartalmaz adatokat.

````python
# Töltsük be az adatokat DataFrame-be!
import pandas as pd
df = pd.read_csv('fifa19.csv.gz')
df
````

````python
# Töröljük az "Unnamed: 0" nevű oszlopot!
del df['Unnamed: 0']
````

````python
# Hány csapat (Club) van az adathalmazban?

# 1. megoldás
len(set(df['Club'].dropna()))
````

````python
# 2. megoldás
df['Club'].nunique()
````

````python
# Milyen nemzetiségű (Nationality) játékosból van a legtöbb az adathalmazban?

# 1. megoldás
df.groupby('Nationality').size().idxmax()
````

````python
df['Nationality'].value_counts().index[0]
````

````python
# Mennyi az átlagéletkor (Age)?
df['Age'].mean()
````

````python
# Kik a legfiatalabb játékosok (az Age oszlop alapján)?
df[df['Age'] == df['Age'].min()]
````

````python
# Ki a legidősebb játékos?
df.loc[df['Age'].idxmax()]
````

````python
# Ki a 10 leggyorsabb játékos a SprintSpeed oszlop alapján?

# 1. megoldás
columns = ['Name', 'Nationality', 'Age', 'SprintSpeed']
df.loc[df['SprintSpeed'].sort_values(ascending=False)[:10].index][columns]
````

````python
# 2. megoldás
df.sort_values('SprintSpeed', ascending=False)[:10][columns]
````

````python
# Ki a leggyorsabb 35 év feletti jétékos?

# 1. megoldás
df[df['Age'] > 35].sort_values('SprintSpeed', ascending=False).head(1)[columns]
````

````python
# 2. megoldás
df.loc[df[df['Age'] > 35]['SprintSpeed'].idxmax()]
````

````python
# Melyik csapatban játszik a legtöbb különböző nemzetiségű játékos?

# 1. megoldás
se = df.groupby(['Club', 'Nationality']).size().reset_index().groupby('Club').size()
se[[se.idxmax()]]
````

````python
# 2. megoldás
df[['Club', 'Nationality']].drop_duplicates()['Club'].value_counts().head(1)
````

````python
# Nemzetiségenként hány játékos található a Fulham csapatában?
df[df['Club'] == 'Fulham']['Nationality'].value_counts()
````

````python
# Mely 10 csapat költi a legtöbbet a játékosok fizetésére (a Wage oszlop alapján)?

def convwage(w):
    w = w[1: -1]
    if w != '': return int(w) * 1000
    else: return 0
    
df['Wage2'] = df['Wage'].apply(convwage)   # Wage oszlop konvertálása számmá
df.groupby('Club')['Wage2'].sum().sort_values(ascending=False)[:10]
````

### Feladatok

- Igazoljuk a ```convwage``` függvény használhatóságát a ```Wage``` oszlop adataira úgy, hogy 'megmutatjuk', hogy az oszlopban csak olyan adat található, amelyre alkalmazható a függvény (és persze helyes eredményt is ad)!
- Módosítsuk a függvényt úgy, hogy a ```Release Clause``` oszlopra is alkalmazható legyen!
- A módosított függvény segítségével készítsünk egy olyan listát, amely a ```Release Clause``` oszlop szerint rangsorolja a 10 'leggazdagabb' klubot!
