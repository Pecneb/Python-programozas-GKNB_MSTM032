### Egyszerű lekérdezések (a [pl.txt](pl.txt) adathalmazon bemutatva)

````python
import pandas as pd
````

````python
# Töltsük be a pl.txt fájl adatait DataFrame-be!
names = ['round', 'hteam', 'ateam', 'hgoals', 'agoals']
df = pd.read_csv('pl.txt', sep='\t', skiprows=6, names=names)
df
````

````python
df.info()
````

````python
df.describe()
````

````python
# "Gólok száma" oszlop felvétele.
df['goals'] = df['hgoals'] + df['agoals']
df
````

````python
# Példák csoportosításra (groupby).
# Melyik fordulóban hány mérkőzés volt?
df.groupby('round').size()
````

````python
# (majdnem) ugyanez, tömörebben
df['round'].value_counts()
````

````python
# Melyik fordulóban hány gól esett?
df.groupby('round')['goals'].sum()
````

````python
# Melyik fordulóban esett a legtöbb gól?
df.groupby('round')['goals'].sum().idxmax()
````

````python
# Melyik fordulóban mennyi volt az átlagos mérkőzésenkénti gólszám?
df.groupby('round')['goals'].mean()
````

````python
# Írjuk ki a tíz leggólgazdagabb fordulót, gólszámokkal együtt!
df.groupby('round')['goals'].sum().sort_values(ascending=False).head(10)
````

````python
# A mérkőzések hány százalékán esett gól?
(df['goals'] > 0).mean() * 100
````

````python
# Melyik mérkőzésen esett a legtöbb gól?
df.loc[df['goals'].idxmax()]
````

````python
# Írjuk ki, hogy a 10., 20. és 30. fordulóban hány gól esett összesen!
df[(df['round'] == 10) | (df['round'] == 20) | (df['round'] == 30)]['goals'].sum()
````

````python
# ...alternatív megoldás:
df[df['round'].isin({10, 20, 30})]['goals'].sum()
````

````python
# ...harmadik megoldás:
df.groupby('round')['goals'].sum()[[10, 20, 30]].sum()
````

````python
# ...negyedik megoldás:
df[(df['round'] % 10 == 0)]['goals'].sum()
````

````python
# Hány gólt rúgott összesen a Manchester United?
df[df['hteam'] == 'Manchester United']['hgoals'].sum() + df[df['ateam'] == 'Manchester United']['agoals'].sum()
````

````python
# alternatív megoldás:
sum([df[df[f'{p}team'] == 'Manchester United'][f'{p}goals'].sum() for p in 'ha'])
````

### Minta vizsgafeladat

Az [unicef.txt](unicef.txt) szövegfájl a világ 5 év alatti népességének élelmezési helyzetéről tartalmaz adatokat. Az egyes sorok felméréseknek felelnek meg, a felmérések országonként időbeli sorrendben vannak felsorolva. Készíts programot, amely beolvassa az unicef.txt tartalmát, minden országra kiválasztja a legfrissebb felmérést, majd a legfrissebb felmérések alapján kiírja az alábbi statisztikákat:

1. Az országok hányadrészében magasabb az Underweight indikátor az Overweight indikátornál?
2. Mely 5 országban a legmagasabb a Severe Wasting indikátor?
3. Régiónként (United Nations Region) mekkora a 20%-nál nagyobb Stunting indikátorral rendelkező országok aránya?
4. Hol a legalacsonyabb a túlsúlyos gyermekek aránya a 10%-nál alacsonyabb Wasting és Stunting indikátorral rendelkező országok között?

````python
# Adatok beolvasása.
df = pd.read_csv('unicef.txt', sep='|', decimal=',')
df
````

````python
df.info()
````

````python
df.describe().T
````

````python
# Legfrissebb felmérés kiválasztása országonként.
df2 = df.groupby('Country').last()
df2
````

````python
# Az országok hányadrészében magasabb az Underweight indikátor az Overweight indikátornál?
(df2['Underweight'] > df2['Overweight']).mean()
# (úgy tekintjük, hogy ha valamelyik érték hiányzik, akkor nem magasabb az Underweight oz Overweight-nél)
````

````python
# Mely 5 országban a legmagasabb a Severe Wasting indikátor?
df2['Severe Wasting'].sort_values(ascending=False).head(5)
````

````python
# Régiónként (United Nations Region) mekkora a 20%-nál nagyobb
# Stunting indikátorral rendelkező országok aránya?
se1 = df2[df2['Stunting'] > 20].groupby('United Nations Region').size()
se2 = df2.groupby('United Nations Region').size()
(se1 / se2).fillna(0)
````

````python
# ...alternatív megoldás:
df2['Stunting>20'] = df2['Stunting'] > 20
df2.groupby('United Nations Region')['Stunting>20'].mean()
````

````python
# Hol a legalacsonyabb a túlsúlyos gyermekek aránya a 10%-nál alacsonyabb
# Wasting és Stunting indikátorral rendelkező országok között?
df2[(df2['Stunting'] < 10) & (df2['Wasting'] < 10)]['Overweight'].idxmin()
````
